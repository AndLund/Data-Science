```python
import pandas as pd
```

```python
df = pd.read_csv('http://dados.ufrn.br/dataset/8bf1a468-48ff-4f4d-95ee-b17b7a3a5592/resource/ff0a457e-76fa-4aca-ad99-48aebd7db070/download/docentes.csv', sep=';')
```

```python
#modifica direto na tabela
df.drop('nome', axis=1, inplace=True)
```

```python
df.describe(include='object')
```

```python
df.sexo.unique()
```

```python
lotacoes = df.lotacao.unique()
```

```python
totalPorLotacao = []
for lot in lotacoes:
    totalPorLotacao.append(df['lotacao'].where(df['lotacao'] == lot).count())
#print(totalPorLotacao)
```

```python
dfl = pd.DataFrame({
    'lotacao': lotacoes,
    'total': totalPorLotacao
})
dfl
```

```python
dfl.sort_values(by=['total'], ascending=False)
```

```python
for lot in lotacoes:
    tf = (df['sexo']).where((df['sexo'] == 'F') & (df['lotacao'] == lot)).count()
    tm = (df['sexo']).where((df['sexo'] == 'M') & (df['lotacao'] == lot)).count()
    tfgb = tf/(tf+tm)*100
    tmgb = tm/(tf+tm)*100
    print(lot)
    print("F: " + str(tfgb) + "%")
    print("M: " + str(tmgb) + "%\n")
    #print(tf)
    #print(tm)
```
